# Read in data

```python
# v for visits
v = pd.read_csv(r'monthly_vis_by_site_zeroed.csv')
v.head()
```

```python
data_geojson = gpd.read_file('EagleCreek_AOI.geojson')
data_geojson
```

```python
geo_vis = data_geojson.merge(v, on=['siteid', 'Name'])
geo_vis['centroid_lon'] = geo_vis['geometry'].centroid.x
geo_vis['centroid_lat'] = geo_vis['geometry'].centroid.y
geo_vis.head()
geo_vis
```

## Sort/Filter based on pre-fire trail popularity

```python
pre_fire_total = v[(v['pre_post_fire'] == 'pre')].groupby(['Name'], as_index=False)['tot_vis'].sum()
pre_fire_total = pre_fire_total.sort_values(by='tot_vis', ascending=False, ignore_index=True)
pre_fire_total['pre_fire_rank'] = pre_fire_total.index + 1
pre_fire_total = pre_fire_total.rename(columns={'tot_vis': 'pre_fire_tot'})
pre_fire_total

geo_vis = geo_vis.merge(pre_fire_total, on='Name')
geo_vis
```

# Build charts

```python
### SELECTIONS

# select trail
multi = alt.selection_multi(
    fields=['siteid'],
    init=[{'siteid': "CRG_008"}],
    empty='none'
)

# hover nearest date
nearest = alt.selection(
        type="single", nearest=True, on="mouseover", fields=["d2p"], empty="none", init={'d2p':1410739200000}
    )

# slider to choose top N in popularity
slider = alt.binding_range(min=1, max=41, step=1, name="Top # Most Popular Trails: ")
topN = alt.selection_single(name='filter', fields=['cutoff'], bind=slider, init={'cutoff': 20})

### LINE CHART

# Draw a rule at the location of the selection
rules = alt.Chart(geo_vis).mark_rule(
        color="black", 
        strokeWidth=1
    ).encode(
        x="d2p",
    ).transform_filter(nearest)

# Helps draw the rules, this is what we are using to hover
selectors = alt.Chart(geo_vis).mark_point().encode(
            x="d2p",
            opacity=alt.value(0), # Make dots invisible. (Can turn opacity to 1 to see what this is doing)
        ).add_selection(nearest)

# This is the main line chart that shows total visits
line_chart = alt.Chart(geo_vis).mark_line().transform_filter(
                   alt.datum.pre_fire_rank <= topN.cutoff
                ).encode(
                    x=alt.X('yearmonth(d2p):T',title='Date'),
                    y=alt.Y('tot_vis:Q',title='Total Visitors'),
                    color = alt.Color('Name:N', scale=alt.Scale(scheme="tableau20"), legend=None),# scale=alt.Scale(range = ['gray', 'gray']), legend=None),
                    strokeWidth=alt.condition(multi, alt.value(4), alt.value(1)),
                    opacity=alt.condition(multi, alt.value(1), alt.value(0.30)),
                ).properties(
                    width=600,
                    height=100,
                    title={
                          "text": ["Outdoor Recreation Before and After the Eagle Creek Fire","What were trends in trail use before and after a major fire in the Columbia Gorge?"],
                          "subtitle": ["", "Pan over the line chart to visualize change in visitors over time", "Hover over each trail area for more info","Click to highlight visitor trends above","Highlight multiple trails with shift+click"],
                          }
                )

line_label_base = alt.Chart(geo_vis).mark_text(align="left").transform_filter(
    "datum.year == 2019 && datum.month == 8"
).encode(
    text="Name",
    x=alt.X('d2p:T'),
    y=alt.Y('tot_vis:Q'),
    opacity=alt.condition(multi, alt.value(1), alt.value(0))
)

# Mark's when Eagle Creek Fire happened
fire_rule = alt.Chart(geo_vis).mark_rule(
        color="orange", 
        strokeWidth=3
    ).encode(
        x="d2p",
    ).transform_filter("datum.year == 2017 && datum.month == 9")

fire_text = alt.Chart(geo_vis).mark_text(
    text = 'Eagle Creek Fire', 
    angle=270,
    fontSize=12
).encode(
        x="d2p",
    ).transform_filter("datum.year == 2017 && datum.month == 10")

# Denote each year
year_rule = alt.Chart(geo_vis).mark_rule(
        color='black',
        strokeDash=[3,5]
    ).encode(
        x="d2p",
    ).transform_filter("[datum.year == 2015 || datum.year == 2016 || datum.year == 2017 || datum.year == 2018 || datum.year == 2019] && datum.month == 1")
    

################## MAP STUFF

# name of current selected trail
trail_title_labels = alt.Chart(geo_vis).mark_text(align='left').encode(
    x = alt.value(0),
    y = alt.value(10),
    text= 'Name',
    size=alt.value(20),
    opacity=alt.condition(multi, alt.value(1), alt.value(0))
)

# Display current date
date_label = alt.Chart(geo_vis).mark_text(align='right').encode(
    x = alt.value(450),
    y = alt.value(10),
    text= 'yearmonth(d2p)',
    size=alt.value(15),
).transform_filter(nearest)

# 
choromap = alt.Chart(geo_vis).mark_geoshape(
        stroke='black',
        strokeWidth=1,
        color = 'black'
    ).encode(
        stroke=alt.Color('burned', 
                         title = 'Burned', 
                         scale=alt.Scale(domain=['no', 'yes'], range=['green', 'orange']), 
                         legend=alt.Legend(
                            orient='bottom',
                             symbolStrokeWidth=5,
                            titleAnchor='middle')
                        ),
        strokeDash=alt.StrokeDash('closed_month', 
                                  title = 'Open/Closed?',
                                  scale=alt.Scale(domain=['open', 'closed']),
                                        legend=alt.Legend(
                                            orient='bottom',
                                            symbolDash=[5,2],
                                             symbolStrokeColor='black',
                                             symbolFillColor='green', # this is required... but not displayed
                                             symbolType='stroke',
                                            titleAnchor='middle')
                                 ),  
        strokeWidth=alt.value(3),
        tooltip=['Name', alt.Tooltip('tot_vis:Q', title='Total visitors'), alt.Tooltip('yearmonth(d2p):T', title='Date'), alt.Tooltip('closed_month', title='Open/Closed') ],
        color=alt.Color("tot_vis:Q", title = 'Total visitors', scale=alt.Scale(domain= [0, 1, 5000, 10000, 15000, 22000], scheme='turbo')) # type='symlog', #range=["white", "white", "green"], 
    ).add_selection(
        multi
    ).transform_filter(nearest).properties(
    width=600,
    height=400,
#     title={
#     "text": " ", 
#     "subtitle": ["Hover over each trail area for more info","Click to highlight visitor trends above","Highlight multiple trails with shift+click"],
#     }
).transform_filter(
    alt.datum.pre_fire_rank <= topN.cutoff
).add_selection(
    topN
)

# mark text/symbol directly on shape of currently selected trail
trail_marker = alt.Chart(geo_vis).mark_text(baseline='top'
 ).encode(
     longitude='centroid_lon:Q',
     latitude='centroid_lat:Q',
     text='Name',
     size=alt.value(16),
     opacity=alt.condition(multi, alt.value(1), alt.value(0))
 ).properties( 
    width=600,
    height=300
)

line_vis = line_chart + selectors + rules + fire_rule + fire_text + year_rule + line_label_base
map_vis = choromap + date_label + trail_marker
final = line_vis & map_vis
final
```
